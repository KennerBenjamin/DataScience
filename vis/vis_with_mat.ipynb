```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
athletes = pd.read_csv("2019_games_athletes.csv")
athletes = athletes[athletes.gender != 'X']
athletes.head()
```

```python
men = athletes.loc[athletes['division']=='Men']
women = athletes.loc[athletes['division']=='Women']
```

```python
sns.countplot(x = 'gender', data = athletes, palette = 'magma')
plt.title('Athletes')
plt.show()
```

```python
sns.boxplot(x="gender", y="age", data=athletes)
```

```python
sns.distplot(men['age'],label='men')
sns.distplot(women['age'],label='women')
plt.legend()
plt.show()
```

```python
g = sns.jointplot(
    data=athletes,
    x="weight", y="height", hue="gender",
    kind="kde",
)
```

```python
sns.displot(data=athletes, x="age",hue="gender")
```

```python
sns.countplot(data=athletes,x="countryoforigincode",order=athletes.countryoforigincode.value_counts().iloc[:5].index)
```
