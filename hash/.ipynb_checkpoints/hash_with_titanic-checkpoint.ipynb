```python
import pandas as pd
import hashlib
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
print(hashlib.algorithms_available)
```

## Hashing the Name with sha512

```python
temp_df = pd.DataFrame()
for index, row in train.iterrows():
    temp_df.loc[index,'hashed_name'] = hashlib.sha512(str(row['Name']).encode('utf-8')).hexdigest()
train = pd.concat([train, temp_df], axis=1, sort=False)
train.drop('Name',axis=1,inplace=True)
```

## Hashing the Ticket with sha3_512

```python
temp_df = pd.DataFrame()
for index, row in train.iterrows():
    temp_df.loc[index,'hashed_ticket'] = hashlib.sha3_512(str(row['Ticket']).encode('utf-8')).hexdigest()
temp_df.head()
train = pd.concat([train, temp_df], axis=1, sort=False)
train.drop('Ticket',axis=1,inplace=True)
```

```python
train.to_csv('hashed.csv')
```
